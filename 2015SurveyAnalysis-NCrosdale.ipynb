````python
#Import packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#Upload data
def load_data(file):                                                    
    return pd.read_csv(file, sep = ",", index_col = False, header=1)
survey_df=load_data("2015 Stack Overflow Developer Survey Responses.csv")
````

Here I  just want to get a broad overview and general info on the data frame

````python
survey_df.info()
````

The goal here was to get the column names so I can easily refer to them as I work with the dataframe

````python
survey_df_column_names = list(survey_df)
print(survey_df_column_names)
````

Here, I wanted to get more information on each column to see if there was any strange data. I didn't see anything strange with the data that I was interested in so I didn't feel the need to clean the data.

````python
#describe each of the variables
survey_df.describe()
````

Looking at the relationship between Age & Industry

````python
survey_df.groupby(
    ['Age', 'Industry']
).count().unstack(level=1)
````

With this information, I can easily see the age range frequency that falls into each industry.

This graph looks at the relationship between the Compensation and Industry

````python
survey_df.groupby(
    ['Compensation','Industry' ]
).count().unstack(level=1)
````

This graph can be very helpful to see where the majority of the respondants make the most or least amount of money.

This bar graph looks at the relationahip between the number of caffinated beverages consumed per day and compensation ranges.

````python
survey_df.groupby(
    ['How many caffeinated beverages per day?','Compensation' ]
).count().unstack(level=1)
````

````python
groupby_caffine = survey_df.groupby(['How many caffeinated beverages per day?']) 
groupby_caffine['Compensation'].value_counts().unstack().plot(
    kind='barh',
  figsize=(8,8)
  )
plt.xlabel('count')
plt.ylabel('Caffinated drinks per day')
````

Here you can see if there is a relationship between caffiene consumption and income. Example: From the data, it looks like most of the people who make more than $160,000 drink about 2 cups per day.

The data below is looking at the relationship between the programming softwares and the compensation they recieve

````python
groupby_system = survey_df.groupby(['Desktop Operating System']) 
groupby_system['Compensation'].value_counts().unstack().plot(
    kind='barh',
  figsize=(8,8)
  )
plt.xlabel('Count')
plt.ylabel('Desktop Used')
````

This was a fun to make a chart out of because I can see which operating systems are most common among individuals in different ranges of compensation. For example, I can see that people who are making $40,000-$60,0000 common use Windows 7 but I can also make the observation that Windows 7 is a common system among all people in the dataframe.

The data below is looking at the relationship between occupation and compensation

````python
groupby_system = survey_df.groupby(['Occupation']) 
groupby_system['Compensation'].value_counts().unstack().plot(
    kind='bar',
  figsize=(15,15)
  )
plt.xlabel('Occupation')
plt.ylabel('Count')
````

In this chart (above), I was interested in looking at the most common compensation of people in certain occupations. I can make the observation that the majority of respondants who are Back-end web developers make less than 20,000 annually.
